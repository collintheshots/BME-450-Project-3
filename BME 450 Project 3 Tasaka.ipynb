```python
# Import CSV File


import pandas as pd

url = 'https://raw.githubusercontent.com/collintheshots/BME-450-Project-3/master/query%20(3).csv'

df = pd.read_csv(url, error_bad_lines=False)

# df.head


# import csv

# dirName = '\\Users\\Art\\Downloads\\CTDdata.csv'

# fileName = pd.read_csv(dirName)

# print(fileName.columns)

# fileName

# with open('\Users\Art\Downloads\CTDdata.csv', newline=',') as csvfile:
```

```python
Time = df.time
Latitude = df.latitude
Longitude = df.longitude
Magnitude = df.mag
```

```python
import matplotlib.pyplot as plt
```

```python
# plot Earthquake Magnitude vs. Time

plt.scatter(Time, Magnitude)
plt.figure(1).set_figheight(8)
plt.figure(1).set_figwidth(20)
plt.title('Earthquake Magnitude vs. Time')
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.show()
```

```python
# Find size of mapped data points to create map

Boundary = (df.longitude.min(), df.longitude.max(),
             df.latitude.min(), df.latitude.max())
Boundary
```

```python
ruh_m = plt.imread('https://github.com/collintheshots/BME-450-Project-3/blob/master/map%20(2).png?raw=true')
```

```python
fig, ax = plt.subplots(figsize = (5,35))



for i in range(0, len(Longitude)):
    
    if(Magnitude[i] > 6):
        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='maroon', s=256)
        
    elif(Magnitude[i] > 5):
        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='red', s=128)
        
    elif(Magnitude[i] > 4):
        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='coral', s=64)
        
    elif(Magnitude[i] > 3):
        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='darkorange', s=32)
        
    elif(Magnitude[i] > 2):
        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='navajowhite', s=16)
        
  #  elif(Magnitude[i] > 1):
#        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='navajowhite', s=8)
        
 #   elif(Magnitude[i] > 0):
#        ax.scatter(Longitude[i], Latitude[i], zorder=1, alpha= .4, c='yellow', s=4)





ax.set_title('Plotting Seismic Data on Map')
ax.set_xlim(Boundary[0],Boundary[1])
ax.set_ylim(Boundary[2],Boundary[3])
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.imshow(ruh_m, zorder=0, extent = Boundary, aspect= 'equal')
```
